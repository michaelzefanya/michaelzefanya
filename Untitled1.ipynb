```python
import pandas as pd

# Load dataset dari file Excel
df = pd.read_excel("diabetes new.xlsx")

# Cek 5 baris pertama
df.head()
```

```python
# Cek info dataset
df.info()

# Cek jumlah missing values
print("\nMissing Values:\n", df.isnull().sum())

# Cek apakah ada data duplikat
print("\nDuplicated Data:", df.duplicated().sum())

# Cek ringkasan statistik dataset
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style seaborn
sns.set_style("whitegrid")

# Plot histogram untuk semua fitur numerik
df.hist(figsize=(12, 8), bins=20)
plt.show()
```

```python
# Hitung korelasi antar fitur
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
print(df.isnull().sum())  # Cek jumlah missing values per kolom
```

```python
print(df.dtypes)
```

```python
print(df.columns)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

df["Cluster"] = clusters  # Tambahkan hasil cluster ke dataset
print(df.head())
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap="viridis")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("K-Means Clustering")
plt.colorbar()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_scaled, clusters)
print(f"Silhouette Score: {score:.4f}")
```

```python
df.to_csv("diabetes_clustered.csv", index=False)
```
